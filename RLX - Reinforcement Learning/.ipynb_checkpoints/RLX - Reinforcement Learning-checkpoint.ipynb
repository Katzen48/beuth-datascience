```python
%matplotlib inline
import gym
from IPython import display
import matplotlib
import matplotlib.pyplot as plt
```

```python
env = gym.make("Taxi-v3")
observationn = env.reset(seed=42)
# env.action_space => Discrete(4) => 0=left, 1=down, 2 = right, 3 = up
# env.observation_space => Discrete(16) => numbered 4x4 box
img = plt.imshow(env.render(mode='rgb_array')) # only call this once
for _ in range(100):
    img.set_data(env.render(mode='rgb_array')) # just update the data
    display.display(plt.gcf())
    display.clear_output(wait=True)
    action = env.action_space.sample() # take random
    # print(action)
    observation, reward, done, info = env.step(action)
    # print(observation)
    # print(reward)
    # print(done)
    # print(info)
    # observation = 0-15 pos, reward = 1.0 or 1.0, 
    # done = true / false, info=prob=0.333
    if done:
        env.reset()
env.close()
```
