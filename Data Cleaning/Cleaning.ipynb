## Age: make all values numeric

```python
df.age = pd.to_numeric(arg=df.age, errors='coerce')
```

```python
df
```

## Age: remove NaN

```python
df = df[df.age.notnull()]
```

```python
df
```

## Age: Remove values under 12 years

Children under 12 years, who would be in the 5-6th grade in Berlin at most, are unlikely to have a email address

```python
df = df[df.age > 12]
```

```python
df
```

## Gender: Remove NaN

```python
df = df[df.gender.notnull()]
```

```python
df
```

## Email: Remove NaN

```python
df = df[df.email.notnull()]
```

```python
df
```

## id: Drop Column

The “id” column is removed, as its data does not seem important for further analysis.

```python
df = df.drop(columns=['id'])
```

```python
df
```

## Drop Duplicates

```python
df = df.drop_duplicates()
```
