```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
import pandas as pd
```

```python
data = pd.read_csv('edlich-kmeans-A0.csv')
data
```

```python
def display_cluster(k):
    kmeans = KMeans(n_clusters=k)
    kmeans = kmeans.fit(data)
    labels = kmeans.predict(data)
    
    C = kmeans.cluster_centers_
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.scatter(data.V1, data.V2, data.V3)
    ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c='#050505', s=1000)
    print(labels)
```

## General Cluster
A visualization of the data

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data.V1, data.V2, data.V3)
```

## k = 3

```python
display_cluster(3)
# First is the clustering vector, second is the graph
```

## k = 4

```python
display_cluster(4)
# First is the clustering vector, second is the graph
```

## k = 5

```python
display_cluster(5)
# First is the clustering vector, second is the graph
```

## k = 6

## k = 7

```python
display_cluster(8)
# First is the clustering vector, second is the graph
```

k = 5 is the best variant.  
It produces good centers for each cluster. Higher values produce more subclusters, which may not necessarily are different groups.
